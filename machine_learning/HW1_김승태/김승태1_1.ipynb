```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.linspace(0, 1, 10) 
y_function = np.sin(2 * np.pi * x)

gaussian_noise = np.random.normal(0, 0.1, size=x.shape)
y_noisy_function = y_function + gaussian_noise

plt.scatter(x, y_noisy_function, label='Gaussian noise', color='none', edgecolors='blue')
plt.plot(np.linspace(0, 1, 100), np.sin(2 * np.pi * np.linspace(0, 1, 100)), color='green', label='sin(2pix)')
plt.title('Function sin(2pix) with a Gaussian noise')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.ylim(-1.5, 1.5)
plt.grid(True)
plt.show()
```
